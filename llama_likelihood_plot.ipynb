```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seed = 41
size = 30

filename = f"../results_csv/ground_truth_probability_{str(seed)}_{str(size)}.csv"
chat_filename = f"../results_csv/chat_ground_truth_probability_{str(seed)}_{str(size)}.csv"
df = pd.read_csv(filename)
chat_df = pd.read_csv(chat_filename)
```

```python
ground_truth = df["ground_truth"]
ground_truth_greedy = df["ground_truth_greedy"]
ground_truth_ns = df["ground_truth_nucleus_sampling"]
greedy = df["greedy"]
ns = df["nucleus_sampling"]
bs = df["beam_search"]
print("without RLHF:")
print("ground_truth:", np.mean(ground_truth), np.var(ground_truth))
print("ground_truth_greedy:", np.mean(ground_truth_greedy), np.var(ground_truth_greedy))
print("ground_truth_nucleus_sampling:", np.mean(ground_truth_ns), np.var(ground_truth_ns))
print("greedy:", np.mean(greedy), np.var(greedy))
print("bs:", np.mean(bs), np.var(bs))
print("ns:", np.mean(ns), np.var(ns))

chat_ground_truth = chat_df["ground_truth"]
chat_ground_truth_greedy = chat_df["ground_truth_greedy"]
chat_ground_truth_ns = chat_df["ground_truth_nucleus_sampling"]
chat_greedy = chat_df["greedy"]
chat_ns = chat_df["nucleus_sampling"]
chat_bs = chat_df["beam_search"]

print("\nwith RLHF:")
print("ground_truth:", np.mean(chat_ground_truth), np.var(chat_ground_truth))
print("ground_truth_greedy:", np.mean(chat_ground_truth_greedy), np.var(chat_ground_truth_greedy))
print("ground_truth_nucleus_sampling:", np.mean(chat_ground_truth_ns), np.var(chat_ground_truth_ns))
print("greedy:", np.mean(chat_greedy), np.var(chat_greedy))
print("bs:", np.mean(chat_bs), np.var(chat_bs))
print("ns:", np.mean(chat_ns), np.var(chat_ns))
```

```python
plt.boxplot([ground_truth, ground_truth_greedy, ground_truth_ns, greedy, ns, bs], labels=['GT', 'GTG', 'GTNS','G', 'NS', 'BS'])
# plt.title('LLaMA2-7b transition probability')
plt.show()

plt.boxplot([chat_ground_truth, chat_ground_truth_greedy, chat_ground_truth_ns, chat_greedy, chat_ns, chat_bs], labels=['GT', 'GTG', 'GTNS','G', 'NS', 'BS'])
# plt.title('LLaMA2-7b-chat transition probability')
plt.show()

```
